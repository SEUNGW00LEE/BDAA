### 필수과제
- bank 데이터 공유 

- bank 데이터와 bank_추가 데이터 
- 이 두 가지 데이터를 병합을 해주시면 됩니다.
- 병합하실 때 어떤 컬럼을 사용했는지와 병합이 잘 되었는지 체크까지 해주세요!
- merge how를  통해서 어떤식으로 붙여지는지 원리를 다 보여주세요 ( inner, right, left, outer) 

```python
import pandas as pd
```

```python
#해당 데이터셋들은 ; 을 기준으로 나눠져있기때문에 sep=';' 인자 추가
bank = pd.read_csv("../dataset/bank.csv", sep=';') 
bank_ad = pd.read_csv("../dataset/bank-additional.csv", sep=';')
```

```python
display(bank)
display(bank_ad)
```

```python
print(bank.shape)
print(bank_ad.shape )
```

```python
pd.merge(bank, bank_ad)
```

##### - 0row인 것을 보아 동일한 data가 겹치지 않은 것을 확인할 수 있음

##### - 위의 shape를 통해 bank_ad에 추가되는 column이 있다는 것을 유추할 수 있음

```python
print(bank.columns)
print(bank_ad.columns)
```

- bank에만 존재하는 칼럼 : 'balance', 'day'
- bank_ad에만 존재하는 칼럼 : 'cons.conf.idx', 'cons.price.idx', 'day_of_week', 'emp.var.rate', 'euribor3m', 'nr.employed'
- 공통칼럼 : 'age', 'job', 'martial', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'duration', 'compaign', 'pdays', 'previous','poutcome', 'y' 

```python
pd.merge(bank, bank_ad, on=['age', 'job'])
#defalt가 inner이므로 , how='inner' 로 적용
```

```python
pd.merge(bank, bank_ad, on=['age', 'job'], how='inner')
#how='inner'을 적지않아도 동일한 결과값 산출
```

```python
pd.merge(bank, bank_ad, on=['age', 'job'], how='outer')
```

- data column이 매우 많으니, 조건을 추가해 데이터를 줄임
- job = unempolyed이고, marital = single 로 추출

```python
df1 = bank.query('job == "unemployed" & marital == "single"')
df2 = bank_ad.query('job == "unemployed" & marital == "single"')
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on=['age','housing'], how='left')
```

```python
pd.merge(df1, df2, on=['age','housing'], how='right')
```
