## 데이터프레임과 시리즈에 대해서 확인하기 !

### 시리즈의 메서드
- append : 2개 이상의 시리즈 연결
- describe : 요약통계
- drop_duplicates : 중복값 없는 시리즈 반환
- equals : 시리즈에 해당 값 가진 요소 확인
- get_values : 시리즈 값 구하기 ( values와 같음)
- isin
- min
- max
- mean
- std, var
- median
- replace : 특정값을 가진 시리즈 값을 교체
- sort_values 정렬
- to_frame

### 과제1 
- 데이터프레임을 직접 만들고 위의 시리즈 메서드를 모두 사용해 주세요!

```python
import pandas as pd

df_epl = pd.DataFrame({'팀' : ['맨시티', '브렌트포드', '토트넘', '풀럼'],
                    '선수':['홀란드','이반 토니','해리 케인', '미트로비치'],
                    '골':[21, 15, 13, 11],
                     '도움':[3, 1, 3, 2]}) 
```

```python
se_goal = df_epl['골']
se_goal
```

```python
se_assist = df_epl['도움']
se_assist
```

```python
se_goal.values #series에 get_values가 아닌 values로 value 값을 가져오는 것 같습니다.
```

```python
### 시리즈의 메서드
print("append:", se_goal.append(se_assist))
print("describe: ", se_goal.describe()) #요약통계
print("duplicates: ",se_goal.drop_duplicates()) #중복값 없는 시리즈 반환
print("equals: ", se_goal.equals(other=se_assist)) # 시리즈에 해당 값 가진 요소 확인
#print(se_goal.get_values()) # 시리즈 값 구하기 ( values와 같음)
print("inin: ",se_goal.isin([15]))
print("min: ",se_goal.min())
print("max: ",se_goal.max())
print("mean: ",se_goal.mean())
print("std: ", se_goal.std())
print("var: ",se_goal.var())
print("median: ",se_goal.median())
print("replace: ", se_goal.replace()) #특정값을 가진 시리즈 값을 교체
print("sort_value : ",se_goal.sort_values()) #정렬
print("to_frame : ",se_goal.to_frame())
```
