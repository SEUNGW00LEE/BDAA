## 2. date_range 함수

날짜 시작일과 종료일 또는 시작일과 기간을 입력하면 범위 내의 인덱스를 생성할 수 있음

```python
pd.date_range('2018-4-1','2018-4-30') #시작일과 종료일
```

```python
pd.date_range(start='2018-4-1',periods=30) #시작일과 기간 입력
```

## freq 인수로 특정한 날짜만 생성할 수 있음

- s:초
- t:분
- H:시간
- D:일
- B:주말이 아닌 평일
- W:일요일
- W-MON:월요일
- M:각 달의 마지막 날
- MS:각 달의 첫날
- BM: 평일 중 각 달의 마지막 날
- BMS: 평일 중 각 달의 첫 날
- WOM-2THU:각 달의 두 번째 목요일
- Q-JAN:각 분기의 첫 달 마지막 날
- Q-DED: 각 분기의 마지막 달의 마지막 날

```python
pd.date_range('2018-4-1','2018-4-30', freq='B')
#2018-4-1부터 2018-4-30까지 평일만 출력
```

```python
pd.date_range('2018-1-1','2018-12-31',freq='W')
#2018년의 일요일만 출력
```

```python
pd.date_range('2018-1-1','2018-12-31',freq='W-Mon')
#2018년의 월요일만 출력
```

# 3. resample 함수

조건으로 주어진 단위를 간격으로 동일하게 데이터를 뽑아냄

```python
crime=pd.read_csv('crime.csv')
```

```python
crime['REPORTED_DATE']=pd.to_datetime(crime['REPORTED_DATE'])
crime_df=crime.set_index('REPORTED_DATE') #시계열데이터를 인덱스로 설정
```

```python
crime_df
```

```python
crime_df.resample('5M').sum()
#5달 간격으로 데이터를 추출한 후 합 구하기
```

# 4. timedelta, offset 함수

timedelta와 offset 함수를 이용하면 날짜 연산 가능

```python
#timedelta 함수 이용하기
from datetime import timedelta
```

```python
n=pd.to_datetime('2021-12-23')
n+timedelta(100) #100일 후 날짜를 계산할 수 있음
```

```python
#offset 함수 이용하기
import pandas.tseries.offsets
from pandas.tseries.offsets import Day, Hour, Second
```

```python
print(n+Day(100)) #100일 후 날짜 계산하기
print(n+Hour(7)) #7시간 후 날짜와 시간 계산하기
print(n+Second(75)) #75초 후 날짜와 시간 계산하기
```

# 5. first, last 필터링 하기

first/last를 통해 시계열데이터에서 처음/끝 값으로부터 특정 기간만큼의 데이터 필터링 할 수 있음

사용법: 데이터명.first(offset) 또는 데이터명.last(offset)


offset에는 기준이 되는 기간이 들어감

```python
sales=pd.read_csv('sales_data.csv',encoding='latin_1') #데이터 불러오기
```

```python
sales['InvoiceDate']=pd.to_datetime(sales['InvoiceDate'])
#날짜 정보열 시계열데이터로 바꾸기
```

```python
sales_=sales.set_index('InvoiceDate') #날짜열을 인덱스로 지정해 시계열데이터로 만들기
```

```python
sales_ #정리된 데이터 확인하기
```

```python
sales_.first('5D') #첫 데이터행으로부터 5일의 데이터 불러오기
```

```python
sales_.last('3M') #마지막 데이터로부터 3달의 데이터 불러오기
```

# 필수분석과제) sales 데이터 분석하기

## 분석 1. 월별로 주문량에 차이가 있는지 확인해보자!!!

```python
df=sales_.loc['2011'] #2011 자료만 가지고 분석 진행
df
```
