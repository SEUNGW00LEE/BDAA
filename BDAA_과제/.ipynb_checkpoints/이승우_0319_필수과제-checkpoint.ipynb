- 범주형 데이터인 전화해지여부 feature도 제외

```python
call_d = call.drop(['ID','전화해지여부'],axis=1)
call_d
```

```python
#데이터 전처리 - 로그변환(log 변환이란 어떤 수치 값을 그대로 사용하지 않고 여기에 로그를 취한 값을 사용하는 것을 말함.)

import numpy as np

call_d['가입일'] = np.log1p(call_d['가입일'])
call_d['음성사서함이용'] = np.log1p(call_d['음성사서함이용'])
call_d
```

```python
#데이터 전처리 - (1) 최대최소 스케일링

from sklearn.preprocessing import MinMaxScaler
#최소최대 스케일링(Min-Max Scaling)은 모든 특성이 정확하게 0과 1사이에 위치하도록 데이터를 정규화한다. 
#. 2차원 데이터셋일 경우 모든 데이터가 x축의 0과 1 y축의 0과 1 사이의 사각 영역에 담기게 된다. 
#정규화 공식은 (데이터 값-최소값)/(최대값-최소값) 이다. 
#Scikit Learn의 MinMaxScaler 함수를 사용해서 구현할 수 있다.
#변환 후의 평균이 정확히 0이 되지 않고 이상치의 악영향을 받기 쉽다는 단점이 있어서 표준화 (Standardization) 방법이 더 자주 쓰인다. 
#한편 이미지 데이터의 픽셀값과 같이 처음부터 0~255로 범위가 정해진 변수는 최소-최대 스케일링을 
#이용하는게 더 자연스러울 수 있다.

MinMax_scaler = MinMaxScaler()

call_MinMax = scaler.fit_transform(call_d)

call_MinMax_df = pd.DataFrame(data=call_MinMax, columns=call_d.columns)
call_MinMax_df
```

```python
call_MinMax_df.hist(figsize=(30,30))
plt.show()
```

```python
#데이터전처리 - (2) 정규화
#표준화(Standardization)는 variance scaling이라고 불리기도 한다. 
#곱셈과 덧셈만으로 변환하는 선형변환을 통해 각 특성의 평균을 0, 분산을 1로 변경하여 모든 특성이 같은 크기를 가지게 한다. 
#그러나 이 방법은 특성의 최소값과 최대값 크기를 제한하지는 않는다. 
#(데이터값-평균)/표준편차로 구하며 이 값을 표준점수 혹은 z-score라고 한다.

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
call_ss = ss.fit_transform(call_d)
call_ss_df = pd.DataFrame(data=call_ss, columns=call_d.columns)
call_ss_df
```

```python
call_ss_df.hist(figsize=(30,30))
plt.show()
```

- 유의미한 차이가 없어보이지만, 최댓값과 최솟값이 정해져있지않은 데이터이므로 
- 최대최소스케일링이 아닌 정규화 방식을 이용한 데이터 전처리를 하겠다.

```python
#drop한 feature 병합
call = pd.concat([call['ID'],call_ss_df,call['전화해지여부']], axis=1)
call
```
